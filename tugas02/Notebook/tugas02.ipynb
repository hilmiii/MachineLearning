```python
import pandas as pd
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# Read the CSV file with a comma delimiter 
df = pd.read_csv('../data/day.csv', sep=',')
```

```python
# Cetak informasi dataset awal
print("Informasi Dataset Awal")
print(f"Jumlah total data: {len(df)}")
print(f"Shape dataset: {df.shape}")
print("\n5 baris data teratas:")
print(df.head())
print("\n" + "="*50 + "\n")
```

```python
# 1. Membagi data: 80% training + 20% testing dari total dataset
df_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
# 2. Membagi data training: 10% validation dari data training
df_train, df_val = train_test_split(df_train, test_size=0.1, random_state=42)
```

```python
# Menampilkan informasi Data Training
print("Data Training")
print(f"Jumlah data training: {len(df_train)}")
print(f"Persentase dari total dataset: {(len(df_train)/len(df))*100:.1f}%")
print("\n5 baris data teratas training:")
print(df_train.head())
print("\n" + "-"*50 + "\n")
```

```python
# Menampilkan informasi Data Validation
print("Data Validation")
print(f"Jumlah data validation: {len(df_val)}")
print(f"Persentase dari total dataset: {(len(df_val)/len(df))*100:.1f}%")
print(f"Persentase dari data training: {(len(df_val)/len(df_train))*100:.1f}%")
print("\n5 baris data teratas validation:")
print(df_val.head())
print("\n" + "-"*50 + "\n")
```

```python
# Menampilkan informasi Data Testing
print("Data Testing")
print(f"Jumlah data testing: {len(df_test)}")
print(f"Persentase dari total dataset: {(len(df_test)/len(df))*100:.1f}%")
print("\n5 baris data teratas testing:")
print(df_test.head())
print("\n" + "="*50 + "\n")
```

```python
# Ringkasan akhir
print("Ringkasan Pembagian Dataset")
print(f"Total dataset: {len(df)} data")
print(f"Training: {len(df_train)} data ({len(df_train)/len(df)*100:.1f}%)")
print(f"Validation: {len(df_val)} data ({len(df_val)/len(df)*100:.1f}%)")
print(f"Testing: {len(df_test)} data ({len(df_test)/len(df)*100:.1f}%)")
print(f"Total setelah pembagian: {len(df_train) + len(df_val) + len(df_test)} data")
```
